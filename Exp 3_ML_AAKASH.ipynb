```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
df = pd.read_csv('Salary_dataset.csv')
df.head()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
model = LinearRegression()
model
```

```python
X= df.drop('Salary', axis=1)
y = df['Salary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,  y_test)
```

```python
modelL1 = Lasso(alpha=10)
modelL1.fit(X_train, y_train)
```

```python
modelL1.score(X_test, y_test)
```

```python
modelL2 = Ridge(alpha=10)
modelL2.fit(X_train, y_train)
```

```python
modelL2.score(X_test, y_test)
```

```python
%matplotlib inline
plt.scatter(X, y)
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.plot(X, model.predict(X), color='red')
plt.plot(X, modelL1.predict(X), color='green', linestyle='--')
plt.plot(X, modelL2.predict(X), color='yellow')
plt.show()
```
